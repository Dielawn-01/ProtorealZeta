## 1. Prime-Harmonic Scanner
Generate resonance data for the T-series lattice.

```python
def run_scan(m_range=50, n_range=50, tiers=[1, 2, 3]):
    results = []
    for k in tiers:
        print(f"Scanning Tier k={k}...")
        for m in range(1, m_range + 1):
            for n in range(1, n_range + 1):
                _, eps, rank = ze.T_m_n(m, n, k)
                results.append({'m': m, 'n': n, 'k': k, 'eps': eps, 'score': 1/(1+eps)})
    return pd.DataFrame(results)

df_res = run_scan()
df_res.to_csv('resonance_data.csv', index=False)
print("Scan complete. Data saved to resonance_data.csv")
```
